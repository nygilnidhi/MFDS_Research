Part 1: Data Loading and Exploration

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Part 2: Data Preprocessing and Feature Engineering

Define Features and Targets for Both Attenuations

```python
# Encode the SYNOP column (weather condition)
data['SYNOPCode'] = LabelEncoder().fit_transform(data['SYNOPCode'])

# Define features and target for RFL_Att (Radio Frequency Link Attenuation)
features = ['Temperature', 'RelativeHumidity', 'RainIntensity', 'Visibility', 'SYNOPCode']
X_rfl = data[features]
y_rfl = data['RFL_Att']  # Target for RF attenuation

# Define features and target for FSO_Att (Free Space Optics Attenuation)
X_fso = data[features]
y_fso = data['FSO_Att']  # Target for Free Space Optics attenuation
```

```python
# Display the selected features and targets
print("\nSelected Features for RFL_Att:")
print(X_rfl.head())
```

```python
print("\nSelected Features for FSO_Att:")
print(X_fso.head())
```

Part 3: Data Splitting

Split Data into Training and Testing Sets

```python
# Split data for RFL_Att
X_train_rfl, X_test_rfl, y_train_rfl, y_test_rfl = train_test_split(X_rfl, y_rfl, test_size=0.2, random_state=42)

# Split data for FSO_Att
X_train_fso, X_test_fso, y_train_fso, y_test_fso = train_test_split(X_fso, y_fso, test_size=0.2, random_state=42)

print("\nData Split Complete:")
print("RFL_Att - Training Set Size:", X_train_rfl.shape)
print("FSO_Att - Training Set Size:", X_train_fso.shape)
```

Part 4: Modeling - Method 1 (Specific Models for Each Weather Condition)

Train Separate Models for Each SYNOP Code

```python
# Dictionaries to store models and predictions for each attenuation type
models_rfl = {}
predictions_rfl = {}

models_fso = {}
predictions_fso = {}

# Train models for RFL_Att
for synop in X_train_rfl['SYNOPCode'].unique():
    X_train_synop = X_train_rfl[X_train_rfl['SYNOPCode'] == synop].drop(columns=['SYNOPCode'])
    y_train_synop = y_train_rfl[X_train_rfl['SYNOPCode'] == synop]

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train_synop, y_train_synop)
    models_rfl[synop] = model

    X_test_synop = X_test_rfl[X_test_rfl['SYNOPCode'] == synop].drop(columns=['SYNOPCode'])
    y_test_synop = y_test_rfl[X_test_rfl['SYNOPCode'] == synop]
    predictions_rfl[synop] = (model.predict(X_test_synop), y_test_synop)

# Train models for FSO_Att
for synop in X_train_fso['SYNOPCode'].unique():
    X_train_synop = X_train_fso[X_train_fso['SYNOPCode'] == synop].drop(columns=['SYNOPCode'])
    y_train_synop = y_train_fso[X_train_fso['SYNOPCode'] == synop]

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train_synop, y_train_synop)
    models_fso[synop] = model

    X_test_synop = X_test_fso[X_test_fso['SYNOPCode'] == synop].drop(columns=['SYNOPCode'])
    y_test_synop = y_test_fso[X_test_fso['SYNOPCode'] == synop]
    predictions_fso[synop] = (model.predict(X_test_synop), y_test_synop)
```

Part 5: Modeling - Method 2 (Generic Models)

Train a Single Model for Each Attenuation Type

```python
# Train generic model for RFL_Att
generic_model_rfl = RandomForestRegressor(n_estimators=100, random_state=42)
generic_model_rfl.fit(X_train_rfl, y_train_rfl)
generic_predictions_rfl = generic_model_rfl.predict(X_test_rfl)

# Train generic model for FSO_Att
generic_model_fso = RandomForestRegressor(n_estimators=100, random_state=42)
generic_model_fso.fit(X_train_fso, y_train_fso)
generic_predictions_fso = generic_model_fso.predict(X_test_fso)
```

Part 6: Evaluation and Visualization

Evaluate Model Performance for Each Method

```python
# Evaluate Method 1 - Specific Models for RFL_Att
for synop, (y_pred, y_true) in predictions_rfl.items():
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"RFL_Att (SYNOPCode = {synop}) - RMSE: {rmse:.2f}, R^2: {r2:.2f}")

# Evaluate Method 2 - Generic Model for RFL_Att
generic_rmse_rfl = np.sqrt(mean_squared_error(y_test_rfl, generic_predictions_rfl))
generic_r2_rfl = r2_score(y_test_rfl, generic_predictions_rfl)
print(f"\nGeneric Model (RFL_Att) - RMSE: {generic_rmse_rfl:.2f}, R^2: {generic_r2_rfl:.2f}")

# Evaluate Method 1 - Specific Models for FSO_Att
for synop, (y_pred, y_true) in predictions_fso.items():
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"FSO_Att (SYNOPCode = {synop}) - RMSE: {rmse:.2f}, R^2: {r2:.2f}")

# Evaluate Method 2 - Generic Model for FSO_Att
generic_rmse_fso = np.sqrt(mean_squared_error(y_test_fso, generic_predictions_fso))
generic_r2_fso = r2_score(y_test_fso, generic_predictions_fso)
print(f"\nGeneric Model (FSO_Att) - RMSE: {generic_rmse_fso:.2f}, R^2: {generic_r2_fso:.2f}")
```

Visualize Results

```python
# Function to plot predictions vs actual values for specific models
def plot_specific_model_predictions(predictions, target_name):
    for synop, (y_pred, y_true) in predictions.items():
        plt.figure(figsize=(8, 6))
        plt.scatter(y_true, y_pred, alpha=0.6)
        plt.xlabel(f"Actual {target_name} Attenuation")
        plt.ylabel(f"Predicted {target_name} Attenuation")
        plt.title(f"Specific Model (SYNOPCode = {synop}) - Predictions vs Actual ({target_name})")
        plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'k--', lw=2)
        plt.show()

# Plot for specific RFL_Att models
plot_specific_model_predictions(predictions_rfl, "RFL_Att")

# Plot for specific FSO_Att models
plot_specific_model_predictions(predictions_fso, "FSO_Att")
```
